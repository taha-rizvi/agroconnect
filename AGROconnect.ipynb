```python
#importing all the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
#importing the dataset
data_path="/commodity_d.csv"
data=pd.read_csv(data_path)
```

```python
from sklearn.model_selection import train_test_split
train,test=train_test_split(data,test_size=0.2,random_state=42)
```

```python
data_new=train.copy()
data_new["Modal Price"]=pd.to_numeric(data_new["Modal Price"],errors='coerce')
data_new["Min Price"]=pd.to_numeric(data_new["Min Price"],errors='coerce')
data_new["Max Price"]=pd.to_numeric(data_new["Max Price"],errors='coerce')
data_new['Commodity Arrivals']=pd.to_numeric(data_new['Commodity Arrivals'],errors='coerce')
data_new['Commodity Traded']=pd.to_numeric(data_new['Commodity Traded'],errors='coerce')
data_new['State']=data_new['State'].astype('category')
data_new['Commodity']=data_new['Commodity'].astype('category')
data_new["APMC"]=data_new["APMC"].astype('category')
data_new.info()
```

```python
data_newsorted=data_new.sort_values(by="Modal Price",ascending=True)
data_newsorted.head()
```

```python
x=data_new["State"]
y=data_new["Commodity"]
fig, ax = plt.subplots()

# Plot scatter plot
scatter = ax.scatter(x, y)

# Set label font size
ax.set_xlabel("State", fontsize=16)
ax.set_ylabel("Modal Price", fontsize=16)
ax.set_title("State vs Modal Price", fontsize=20)
fig.set_size_inches(30, 5)
# Adjust layout and display plot
plt.tight_layout()
plt.show()

```
